```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('sample.csv')
data.tail() # Display the last few rows of the DataFrame
```

```python
plt.scatter(data.age, data.job)
```

```python
x = data[['age']]
x
```

```python
y = data['job']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression # logisticRegression kiyanne class ekak
```

```python
model = LogisticRegression()# create an object of LogisticRegression class
```

```python
model.fit(x_train, y_train) # fit method eka use karanne model ekata training data eka dena eka
```

```python
model.predict(x_test)
```

```python
x_test
```

```python
y_test
```

```python
model.score(x_test, y_test) # accuracy eka pennanne score method eka use karala
```

```python
age = [[28]]
model.predict(age) # meka use karanne model ekata new data ekak dena eka prediction ekak ganna
```
